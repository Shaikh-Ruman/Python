```python
#1 Load & Explore the dataset of 'CAR CRASHES'
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = sns.load_dataset('car_crashes')
df.head()
```

```python
#2 size of dataset
df.shape
```

```python
#3 data type of car crashes
df.dtypes
```

```python
#4 Check for missing values
df.isnull()
```

```python
#4 Total missing values
df.isnull().sum()
```

```python
#5 Check for unique values in a column
unique_values ={'total': df['total'].unique(),
                'speeding': df['speeding'].unique(),
                'alcohol':df['alcohol'].unique(),
                'not_distracted':df['not_distracted'].unique(),
                'no_previous': df['no_previous'].unique(),
                'ins_premium': df['ins_premium'].unique(),
                'ins_losses': df['ins_losses'].unique()}
unique_values
```

```python
#6) Average percent of speeding related accidents
avg_percent = df['speeding'].mean()
print(avg_percent)
```

```python
#7) Maximum no of car crashes
max_crash_state = df.loc[df['total'].idxmax()]
print(f"The state with the maximum number of car crashes is {max_crash_state['abbrev']} with {max_crash_state['total']} crashes.")
```

```python
#8) alcolhol related 30% & above
high_alcohol = df[df['alcohol']>30].shape[0]
print(f"There are {high_alcohol} states with alcohol related crashes above 30%.")
```

```python
#9) new column - total per million
df['total_per_million'] = df['total'] * 1000000
df.head()
```

```python
#10) Scatter plot - alcohol % & total crashes

plt.figure(figsize=(8,5))
sns.scatterplot(data = df, x = 'alcohol', y = 'total', color = 'blue')
plt.title('Scatter Plot of Alcohol % vs Total Crashes')
plt.xlabel('Alcohol %')
plt.ylabel('Total Crashes')
plt.grid(True)
plt.show()
```

```python
#11) Bar plot - 5 states with speeding percentages

top_speeding = df.nlargest(5, 'speeding')
plt.figure(figsize=(10, 6))
sns.barplot(data=top_speeding, x='abbrev', y='speeding')
plt.title('Top 5 States by Speeding-Related Crashes')
plt.xlabel('State')
plt.ylabel('Speeding-Related Crashes (%)')
plt.show()
```

```python
#12) Heat map - Correlation of all umerical values

plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(),
            annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Car Crash Variables')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'float64'])
correlation_matrix = numeric_df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',linewidth=0.6)
plt.title('Correlation Heatmap of Car_crashes Data')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['abbrev'],df['speeding'], marker='s', linestyle ='--',color='green', label='abbrev')
plt.title("Total speeding car crashes", fontsize=18) # Adds title to lineplot
plt.xlabel("abbrev", fontsize=16) #adds x - axis label
plt.ylabel("Speeding", fontsize=16) #adds y-axis label
plt.legend(loc='upper left') # position of legend
plt.grid(True,linestyle='-',alpha=0.8) #adds grid , alpha is the opacity of line
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(df['alcohol'], labels=df['abbrev'], autopct='%1.1f%%', startangle=45)
plt.title("Alcohol car crashes")
plt.show()
```
