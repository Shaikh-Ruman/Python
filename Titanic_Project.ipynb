```python
# Titanic Dataset

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Q1: Load the Titanic dataset from Seaborn
df = sns.load_dataset('titanic')
print(df.head())

# Q2: Check for missing values in each column
print("\nMissing Values:\n", df.isnull().sum())

# Q3: Drop columns that are not useful for this analysis
df = df.drop(columns=['deck', 'embark_town', 'alive', 'class', 'who'])

# Q4: Fill missing 'age' values with the mean age
df['age'] = df['age'].fillna(df['age'].mean())

# Q5: Drop rows where 'embarked' is missing
df = df.dropna(subset=['embarked'])

# Q6: Display unique values in 'sex' and 'embarked'
print("\nUnique 'sex' values:", df['sex'].unique())
print("Unique 'embarked' values:", df['embarked'].unique())

# Q7: Count number of male and female passengers
print("\nPassenger count by gender:\n", df['sex'].value_counts())

# Q8: Use a loop to print all numeric columns
print("\nNumeric columns:")
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        print("-", col)

# Q9: Count how many passengers were over 60 years old
count = 0
for age in df['age']:
    if age > 60:
        count += 1
print("\nPassengers over 60 years old:", count)


# Q11: Create a histogram of the 'age' column
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Q13: Calculate the overall survival rate
print("\nOverall Survival Rate:", df['survived'].mean())

# Q14: Find survival rate for each gender
print("\nSurvival rate by gender:\n", df.groupby('sex')['survived'].mean())

# Q15: Find average age of survivors and non-survivors
print("\nAverage age by survival:\n", df.groupby('survived')['age'].mean())

# Q16: Determine which 'embarked' location had the highest survival rate
print("\nSurvival rate by embarkation:\n", df.groupby('embarked')['survived'].mean())

# Q17: Print passengers under 10 years old who survived
print("\nChildren under 10 who survived:")
print(df[(df['age'] < 10) & (df['survived'] == 1)][['sex', 'age', 'survived']])


# Q19: Which combination of gender and class had the lowest survival rate?
print("\nSurvival rate by gender and class:\n", df.groupby(['sex', 'pclass'])['survived'].mean())
```
