```python
import numpy as np
import matplotlib.pyplot as plt

category = ['P','Q','R','S','T']
values = [5,15,20,10,25]

#Create a bar chart
plt.bar(category, values, color = 'blue', alpha = 0.8)

#Add titles and labels
plt.title('Basic bar chart')
plt.xlabel('Category')
plt.ylabel('Values')

plt.show()
```

```python
data = np.random.randn(1000)

#Create a histogram
plt.hist(data, bins = 30, color = 'pink', edgecolor = 'blue', alpha = 0.5)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')

#Line Plot
sns.lineplot(x = 'total_bill', y = 'tip', data = tips)
plt.title('Line Plot')
plt.show()

#Bar Plot
sns.barplot(x = 'day', y = 'total_bill', data = tips, palette = 'viridis')
plt.title('Bar Plot')
plt.show()

#Scatter Plot
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'day', style = 'time', palette = 'coolwarm')
plt.title('Scatter Plot')
plt.show()

#Histogram
sns.histplot(data = tips['total_bill'], bins = 20, kde = True, color = 'purple')
plt.title('Histogram')
plt.show()

#Pairplot
sns.pairplot(tips, hue = 'day', palette = 'pastel')
plt.title('Pairplot')
plt.show()
```

```python
#Bar chart using Plotly

import plotly.express as px
import pandas as pd

#Sample data
data = {'State':['Odisha','Haryana','Gujrat','Maharashtra'],
        'Sales':[35000, 40000, 28000, 32000]}
df = pd.DataFrame(data)

#Barchart
fig = px.bar(df, x = 'State', y = 'Sales', color = 'State', title = 'Sales by State')
fig.show()

#Lineplot
fig = px.line(df, x = 'State', y = 'Sales', title = 'Sales Over Time')
fig.show()

#Scatterplot
fig = px.scatter(df, x = 'State', y = 'Sales', color = 'State', title = 'Sales by State')
fig.show()
```
